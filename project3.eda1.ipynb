### <span style="color:lightblue"><b>Date Column Identification and Datetime Conversion</b></span>

Date and time information often plays a significant role in data analysis because it enables the study of **trends, seasonality, temporal patterns, and time-based behavior**. However, date variables are frequently imported as text (`object`) rather than proper datetime formats, limiting their analytical usefulness.

To address this issue, the project automatically identifies potential date-related columns and converts them into **datetime format** using Pandas.

Automated datetime conversion improves preprocessing efficiency and ensures that temporal variables are correctly formatted for advanced analysis and visualization.

This step contributes to a more structured and intelligent preprocessing workflow.

### <span style="color:lightblue"><b>Objectives of Datetime Conversion</b></span>

- **Identify date-related variables automatically**
- **Convert textual dates into datetime format**
- **Improve temporal data usability**
- **Support trend and time-series analysis**
- **Reduce manual preprocessing effort**
- **Prepare data for advanced analytics**

```python
# DATE DETECTION

possible_date_cols = []

for col in df.columns:

    if 'date' in col.lower() or 'time' in col.lower():

        possible_date_cols.append(col)
```

```python

for col in possible_date_cols:

    try:

        df[col] = pd.to_datetime(
            df[col],
            errors='coerce'
        )

        df[f'{col}_year'] = df[col].dt.year
        df[f'{col}_month'] = df[col].dt.month
        df[f'{col}_day'] = df[col].dt.day

    except:
        pass
```

By examining variables independently, analysts can uncover meaningful insights regarding data structure, detect anomalies, and evaluate overall feature quality prior to conducting more complex multivariate investigations or predictive modeling.

### <span style="color:lightblue"><b>OUTLIER DETECTION</b></span>

```python
# IQR Method
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

    print(f"\n{col}")
    print("Number of Outliers:", outliers.shape[0])
```
